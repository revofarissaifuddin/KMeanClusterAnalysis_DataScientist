remove Outliers Data

```python
#remove outliner
def remove_outliner_IQR(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3-Q1
    DF_F = df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR)))]
    return DF_F
df_out_r = remove_outliner_IQR(df[['Negara','Kematian_anak','Ekspor', 'Kesehatan',
                                  'Impor', 'Pendapatan', 'Inflasi',
                                  'Harapan_hidup', 'Jumlah_fertiliti', 'GDPperkapita']])
df_out_r.dropna(axis=0,inplace=True)
df_out_r1 = remove_outliner_IQR(df_out_r[['Negara','Kematian_anak','Ekspor', 'Kesehatan',
                                  'Impor', 'Pendapatan', 'Inflasi',
                                  'Harapan_hidup', 'Jumlah_fertiliti', 'GDPperkapita']])
df_out_r1.dropna(axis=0,inplace=True)
df_out_r2 = remove_outliner_IQR(df_out_r1[['Negara','Kematian_anak','Ekspor', 'Kesehatan',
                                  'Impor', 'Pendapatan', 'Inflasi',
                                  'Harapan_hidup', 'Jumlah_fertiliti', 'GDPperkapita']])
df_out_r2.dropna(axis=0,inplace=True)
df_out_r3 = remove_outliner_IQR(df_out_r2[['Negara','Kematian_anak','Ekspor', 'Kesehatan',
                                  'Impor', 'Pendapatan', 'Inflasi',
                                  'Harapan_hidup', 'Jumlah_fertiliti', 'GDPperkapita']])
df_out_r3.dropna(axis=0,inplace=True)
df_out_r3.fillna(method ='ffill')
df_out_r3.isnull().sum()
print('Nilai Maximum dari masing-masing Variable adalah: ')
print(df_out_r3.describe())
display(df_out_r3)
dfNew = df_out_r3['Negara']
dfNew
```

Elbow graph

```python
sc=StandardScaler()
df_out_r3.drop('Negara', axis=1, inplace=True)
sc.fit_transform(df_out_r3.astype(float))
```

```python
#Elbow method untuk menemukan jumlah k kluster yang tepat
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_out_r3)
    wcss.append(kmeans.inertia_)
plt.subplots(figsize=(10,8))
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')

plt.show()
```
